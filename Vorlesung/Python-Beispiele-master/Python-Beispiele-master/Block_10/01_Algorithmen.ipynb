# Algorithmen
Ein Algorithmus ist eine in der Beschreibung und Ausführung <b>endliche, deterministische und eindeutige</b> Vorschrift zur Lösung eines Problems, die <b>effizient</b> sein sollte.

### Aufgabe 1
In Worträtseln sind in einem rechteckigen Feld Worte versteckt, die horizontal, vertikal oder diagonal (auch rückwärts) geschrieben sind. In einem kleinen Feld lassen sich gesuchte Worte noch schnell selbst finden. Schon bei 50 x 20 Zeichen ist es aber sehr mühsam.<br>
Entwickeln Sie einen Algorithmus, der in einem beliebigen Feld von Buchstaben die Wörter:<br>
MASCHINENBAU, INFORMATIK, HALLO,
PYTHON, OSTFALIA<br>
findet und die Position des Anfangsbuchstabens und die Richtung des Worts als Himmelsrichtung (N, O, S, W, NO, NS, SW, NO) ausgibt.

### Aufgabe 2
In der Datei "Wortsalat.txt" sind diese Wörter in einem 8x20 Feld versteckt. Lesen Sie die Datei in eine geeignete Datenstruktur ein. Implementieren Sie dann den in Aufgabe 1 entwickelten Algorithmus.
